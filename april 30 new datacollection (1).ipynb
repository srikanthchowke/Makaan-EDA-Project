```python
import requests
from bs4 import BeautifulSoup
import re
import numpy as np
import pandas as pd
```

```python
builders=[]
size=[]
property_type=[]
property_address=[]
city =[]
area=[]
rate_sqft=[]
price=[]
construction_status=[]
listed_by=[]
for i in range(1, 21):
    print("https://www.makaan.com/listings?beds=1,2,3,3plus&baths=any&propertyType=apartment,villa,independent-house,studio-apartment&sortBy=relevance&listingType=buy&pageType=LISTINGS_PROPERTY_URLS&cityName=Hyderabad&cityId=12&templateId=MAKAAN_CITY_LISTING_BUY&sellerRating=any&page="+str(i)+"&_=1714450864718")
    url = ("https://www.makaan.com/listings?beds=1,2,3,3plus&baths=any&propertyType=apartment,villa,independent-house,studio-apartment&sortBy=relevance&listingType=buy&pageType=LISTINGS_PROPERTY_URLS&cityName=Hyderabad&cityId=12&templateId=MAKAAN_CITY_LISTING_BUY&sellerRating=any&page="+str(i)+"&_=1714450864718")
    page = requests.get(url)

    soup = BeautifulSoup(page.text)

    # builders
    build = soup.find_all("div", class_="seller-info")

    for i in build:
        builders.append(i.text)

    # size
    proptype = soup.find_all("div", class_="title-line-wrap")

    for i in proptype:
        a = i.text
        b = re.findall(("(\d+\s\w+)"), a)
        if b:
            size.append(b[0])

    # property type
    for i in proptype:
        a = i.text
        b = re.findall(("\d+\s\w+\s(\w+)"), a)
        if b:
            property_type.append(b[0])

    # property address
    #<span itemprop="addressLocality"><strong>Shaikpet</strong></span>
    loc = soup.find_all("span",itemprop="addressLocality")
    for i in loc:
        a = i.text
        b = re.findall(("\w+"), a)
        if b:
            property_address.append(b[0])

    # city
    #<span class="cityName" itemprop="addressRegion">Hyderabad</span>
    loc = soup.find_all("span",class_="cityName")
    for i in loc:
        a = i.text
        b = re.findall("\w+",a)
        if b:
            city.append(b[0])

    # AREA
    ar = soup.find_all("td", class_="size")

    for i in ar:
        area.append(i.text)

    # rate sqft
    persqt = soup.find_all("td", class_="lbl rate")

    for i in persqt:
        rate_sqft.append(i.text.split(" /")[0])

    # price
    p = soup.find_all("td", class_="price")

    for i in p:
        price.append(i.text)

    # construction status
    constatus = soup.find_all("td", class_="val")
    for i in constatus:
        construction_status.append(i.text)

    # listed by
    listed = soup.find_all("span", class_="seller-type")

    for i in listed:
        listed_by.append(i.text)
```

```python
dat_hyd=   {"BUILDERS":builders,
            "SIZE":size,
            "PROPERTY_TYPE":property_type,
            "PROPERTY_ADDRESS":property_address,
            "CITY":city,
            "AREA":area,
            "RATE_SQFT":rate_sqft,
            "PRICE":price,
            "CONSTRUCTION_STATUS":construction_status,
            "LISTED_BY":listed_by}
```

```python
print(len(builders))
print(len(property_type))
print(len(property_address))
print(len(city))
print(len(area))
print(len(rate_sqft))
print(len(price))
print(len(construction_status))
print(len(listed_by))
```
