## 3.trends of most common property types in different addresses within city

```python
prop_trends=df.groupby(["CITY","PROPERTY_TYPE"])["PROPERTY_ADDRESS"].apply(lambda x: x.value_counts().sort_values(ascending=False).head(5))
prop_trends
```

```python
prop_trends = prop_trends.reset_index()
plt.figure(figsize=(12, 6))

sns.barplot(data=prop_trends, x='CITY', y='PROPERTY_ADDRESS', hue='PROPERTY_TYPE', palette='Set2')
plt.title('Top Property Type Trends by City')
plt.xlabel('City')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.legend(title='Property Type', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.show()
```
